a cumulative distribution function for the normal distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_file = open("data/iris.csv", "r")
df = pd.read_csv(data_file, delimiter=";", decimal=",")
print(df)
```

```python
def normal_dist(x , mean , sd):
    return((np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2))
```

```python
sl = df["sepal_length"]
sw = df["sepal_width"]
pl = df["petal_length"]
pw = df["petal_width"]
```

```python
n_sl = normal_dist(sl, sl.mean(), sl.std())
print(n_sl)
plt.plot(sl, n_sl)
plt.xlabel('Data points')
```
